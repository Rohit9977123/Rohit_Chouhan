Q2.In which year highest number of the TV shows & Movies were raleased ? show with Bar Graph

```python
df.head(2)
```

to_datetime

```python
import matplotlib.pylab as plt
df["year_n"] = pd.to_datetime(df["Release_Date"].str.strip(), format="%B %d, %Y")


df["year"]=df["year_n"].dt.year

c=df["year"].value_counts().reset_index()
plt.figure(figsize=(12,6))
ax=sns.barplot(data=c,x="year",y="count")
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

Q. 10) How can we sort the dataset by Year ?

```python
df.head(2)
```

```python
xdf=df.dropna(subset=["year"])
c=xdf["year"].sort_values(ascending=False).reset_index()
c=c.rename(columns={"index":"values"})
plt.figure(figsize=(12,6))
sns.barplot(data=c,x="year",y="values")
plt.show()
```
